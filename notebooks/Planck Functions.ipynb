##Planck function in wavelength

```python
# Solar
#wv = arange(.1,10.,0.01)*1e-6
#bt = 6000.
# Terrestrial
wv = arange(2.,100.,0.01)*1e-6
bt = 288.

h = 6.6260755e-34  # Planck constant in Js
c = 2.99792458e8   # photon speed in m/s
k = 1.380658e-23   # Boltzmann constant in J/K
radiance = ( (2*h*c*c) * 1e-6 ) / ( wv**5 * ( np.exp((h*c)/(k*wv*bt)) - 1 ) )
```

```python
figure()
#plot(wv/1e-6,radiance)
loglog(wv/1e-6,radiance)
xlabel('Wavelength (um)')
ylabel('Radiance [W m-2 sr-1 um-1]')
title('Wavelength versus Radiance')
```

##Planck function in wavenumber

```python
wn = arange(50., 2000., 1.)
bt = 288.

h = 6.6260755e-34  # Planck constant in Js
c = 2.99792458e8   # photon speed in m/s
k = 1.380658e-23   # Boltzmann constant in J/K
C1 = 2*h*c*c*1e8
C2 = h*c/k*1e2
radiance = 1000. * C1 * (wn*wn*wn) / (np.exp((C2*wn)/bt)-1)
```

```python
plot(wn,radiance)
xlabel('Wavenumber (cm-1)')
ylabel('Radiance [mW m-2 sr-1 (cm-1)-1]')
title('WaveNUMBER versus Radiance')
```
