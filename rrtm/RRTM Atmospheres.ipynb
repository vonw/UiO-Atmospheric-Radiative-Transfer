# RRTM Model Atmospheres 

### This notebook graphs the vertical profiles of temperature and trace gases in the model atmospheres used in the  Rapid Radiative Transfer Model (RRTM).

```python
%pylab inline
```

```python
ars = loadtxt('ars.dat')
sas = loadtxt('sas.dat')
mls = loadtxt('mls.dat')
trp = loadtxt('trp.dat')
```

```python
ars[0]
```

## Temperature

```python
figure(figsize=(8,10))
plot(ars[:,3]-273.15,ars[:,1])
plot(sas[:,3]-273.15,sas[:,1])
plot(mls[:,3]-273.15,mls[:,1])
plot(trp[:,3]-273.15,trp[:,1])
grid()
xlabel('Temperature (C)')
ylabel('Altitude (km)')
title('Temperature Profiles for RRTM')
legend(('Arctic Summer','Sub-arctic Summer','Mid-latitude Summer','Tropical'),loc='best')
```

## Water Vapor

```python
figure(figsize=(8,10))
semilogx(ars[:,6],ars[:,1])
semilogx(sas[:,6],sas[:,1])
semilogx(mls[:,6],mls[:,1])
semilogx(trp[:,6],trp[:,1])
grid()
xlabel('Water Vapor Concentration (ppmv)')
ylabel('Altitude (km)')
title('Temperature Profiles for RRTM')
legend(('Arctic Summer','Sub-arctic Summer','Mid-latitude Summer','Tropical'),loc='best')
```
